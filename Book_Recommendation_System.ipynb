# [3] Popularity Recommender

# [3.1] calculate recommendations depending on calculated rating

* Aggregate ratings table on ISBN and calculate mean & count

```python
def Aggregate_on_ISBN():
  populartiy = ratings.groupby('ISBN').agg({'Book-Rating': ['mean', 'count']}).reset_index()
  populartiy.columns = ['ISBN', 'ratings_mean', 'ratings_count']
  populartiy.sort_values(by='ratings_mean', ascending=False)
  return populartiy
```

calculate predicted Rating using the formula: (v/(v+m) * R) + (m/(m+v) * C) where:

*  v is the number of ratings for the book;

*  m is the minimum rating to say this book is good

*  R is the average rating of the book

*  C is the mean ratings across the whole books

```python
def weighted_rating(x, m, C):
  v = x['ratings_count']
  R = x['ratings_mean']
  return (v/(v+m) * R) + (m/(m+v) * C)
```

* calculate predicted ratings for books and sort them in descending order

```python
def calculate_Popularity_Rating():
  pop = Aggregate_on_ISBN()
  C = pop['ratings_mean'].mean()
  pop['predicted_rating'] = pop.apply(lambda x: weighted_rating(x, 8, C), axis=1)
  pop = pop[pop['ISBN'].isin(test['ISBN'])]
  ret = pop.sort_values('predicted_rating', ascending = False)
  return ret
Pop_rec = calculate_Popularity_Rating()
```

```python
def Popularity_Recommendations():
  return Pop_rec
```

# [4.2] First Content Based Recommender:  **Book Author Recommneder** 

* Preparing Data

```python
train_author = pd.merge(train, books, how="left", on="ISBN")
test_author = pd.merge(test, books, how="left", on="ISBN")
train_author.head(5)
```

* Calculating **Predicted Rating** on train dataset for each **author**.
  **Predicted Rating** is a combination between **ratings mean** for a specific author and **ratings count** for this author

```python
def normalize(value, old_max, old_min, new_max=10.0, new_min=0.0):
  old_range = (old_max - old_min)  
  new_range = (new_max - new_min)  
  return (((value - old_min) * new_range) / old_range) + new_min
```

```python
def Book_Author_Rating_Calculation():
  res = train_author.groupby(['User-ID','Book-Author']).agg({'Book-Rating': ['mean', 'count']}).reset_index()
  res.columns = ['User-ID', 'Book-Author' , 'ratings_mean', 'ratings_count']
  res = res.sort_values(by=[ 'User-ID', 'ratings_count'], ascending=[True,False])
  res2 = res.groupby('User-ID').agg({'ratings_count':['max','min']}).reset_index()
  res2.columns = ['User-ID','max','min']
  res = pd.merge(res, res2, how="left", on="User-ID")
  res['ratings_count'] = res[['max','min','ratings_count']].apply(lambda x: normalize(x['ratings_count'],x['max']+1,x['min']-1) , axis=1)
  res['predicted_rating'] = (res['ratings_count'] + res['ratings_mean']*2)/3
  res = res.sort_values(by=[ 'User-ID','predicted_rating', 'Book-Author' ,], ascending=[True,False,True])
  return res
Book_Auth_Rec = Book_Author_Rating_Calculation()
```

* author-based Recommender: for a specific user we extract favourite authors from above calculations and try to find books which is written by favourite authors in test dataset 

```python
def Book_Author_Recommendations(userId):
  Best_Authors = Book_Auth_Rec[Book_Auth_Rec['User-ID'] == userId][['Book-Author','predicted_rating']]
  rec_books = pd.merge(Best_Authors, test_author, how="left", on="Book-Author")
  rec_books = rec_books.dropna()
  return rec_books[['ISBN','predicted_rating']]
```

# [4.3] Second Content Based Recommender: **Publisher Recommender**


* Preparing Data

```python
train_publisher = pd.merge(train, books, how="left", on="ISBN")
test_publisher = pd.merge(test, books, how="left", on="ISBN")
train_publisher.head(5)
```

* Calculating **Predicted Rating** on train dataset for each **publisher**.
  **Predicted Rating** is a combination between **ratings mean** for a specific publisher and **ratings count** for this publisher

```python
def Publisher_Rating_Calculation():
  res = train_publisher.groupby(['User-ID','Publisher']).agg({'Book-Rating': ['mean', 'count']}).reset_index()
  res.columns = ['User-ID', 'Publisher' , 'ratings_mean', 'ratings_count']
  res = res.sort_values(by=[ 'User-ID', 'ratings_count'], ascending=[True,False])
  res2 = res.groupby('User-ID').agg({'ratings_count':['max','min']}).reset_index()
  res2.columns = ['User-ID','max','min']
  res = pd.merge(res, res2, how="left", on="User-ID")
  res['ratings_count'] = res[['max','min','ratings_count']].apply(lambda x: normalize(x['ratings_count'],x['max']+1,x['min']-1) , axis=1)
  res['predicted_rating'] = (res['ratings_count'] + res['ratings_mean']*2)/3
  res = res.sort_values(by=[ 'User-ID','predicted_rating', 'Publisher' ,], ascending=[True,False,True])
  return res
Publish_Rec = Publisher_Rating_Calculation()
```

* Publisher-based Recommender: for a specific user we extract favourite publishers from above calculations and try to find books which is published by favourite publishers in test dataset 

```python
def Publisher_Recommendations(userId):
  Best_Publishers = Publish_Rec[Publish_Rec['User-ID'] == userId][['Publisher','predicted_rating']]
  rec_books = pd.merge(Best_Publishers, test_publisher, how="left", on="Publisher")
  rec_books = rec_books.dropna()
  return rec_books[['ISBN','predicted_rating']]
```

# [5] Collaborative Recommender

* **The total size of ratings data is too large so pivot method could not run and end up with an error because of data size.** here we only apply on test data which we'll recommend books from.


```python
users_items_pivot_matrix_df = test.pivot(index='User-ID', columns='ISBN', values='Book-Rating').fillna(0)
users_items_pivot_matrix_df.head(10)
```

```python
users_items_pivot_matrix = users_items_pivot_matrix_df.values
users_items_pivot_matrix
```

```python
from scipy.sparse.linalg import svds

#The number of factors to factor the user-item matrix.
NUMBER_OF_FACTORS_MF = 15
#Performs matrix factorization of the original user item matrix
#U, sigma, Vt = svds(users_items_pivot_matrix, k = NUMBER_OF_FACTORS_MF)
U, sigma, Vt = svds(users_items_pivot_matrix, k = NUMBER_OF_FACTORS_MF)
```

```python
import numpy as np
sigma = np.diag(sigma)
```

```python
sigma.shape
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) 
all_user_predicted_ratings
```

```python
preds_df = pd.DataFrame(all_user_predicted_ratings, columns=users_items_pivot_matrix_df.columns, index=users_items_pivot_matrix_df.index)
preds_df.head(10)
```

```python
def Collaborative_Recommendations(userId):
  sorted_user_predictions = preds_df.loc[userId].sort_values(ascending=False)
  recommendations = {'ISBN': sorted_user_predictions.index, 'predicted_rating': sorted_user_predictions.values}
  recommendations_df = pd.DataFrame(recommendations)
  return recommendations_df
```

# [6] Hybrid Recommender

```python
def Hybrid_Recommender(userId):
  Rec1 = Book_Author_Recommendations(userId)
  Rec2 = Collaborative_Recommendations(userId)
  mx = Rec2['predicted_rating'].max()
  mn = Rec2['predicted_rating'].min()
  Rec2['predicted_rating'] = Rec2['predicted_rating'].apply(lambda x: normalize(x,mx,mn))
  Rec = pd.merge(Rec1,Rec2,how="outer",on="ISBN").fillna(0)
  Rec['predicted_rating'] = (Rec['predicted_rating_x']+Rec['predicted_rating_y'])/2
  Rec = Rec.drop(['predicted_rating_x','predicted_rating_y'],axis=1)
  return Rec
```

# [7] General Recommender
this is a general recommender which combine all recommenders in one place

```python
def get_Recommendations(ModelNumber,userId):
  if ModelNumber == 1:
    return Popularity_Recommendations()
  elif ModelNumber == 2:
    return Book_Author_Recommendations(userId)
  elif ModelNumber == 3:
    return Publisher_Recommendations(userId)
  elif ModelNumber == 4:
    return Collaborative_Recommendations(userId)
  elif ModelNumber == 5:
    return Hybrid_Recommender(userId)
```

# [8.1] Evaluation Function

* Getting from test data (unseen data) the actual book preferences for a specific user

```python
def sort_and_threshold_test_set(thre = 8):
  ret = test[(test['Book-Rating'] >= thre)].sort_values(by='Book-Rating', ascending=False)
  return ret
sorted_thresholded_test_set = sort_and_threshold_test_set()
```

```python
def actual_favourite_for_user(userId):
  favorites = sorted_thresholded_test_set[(sorted_thresholded_test_set['User-ID'] == userId)]['ISBN']
  return favorites
```

* Getting recommendation thresholded on some predicted rating threshold

```python
def get_top_recommended_N(N,recommended):
  #rec = recommended[(recommended['predicted_rating'] >= thre)].sort_values(by='predicted_rating', ascending=False)
  return recommended.head(N)
```

* Evalure the recommendations for a specific user by calculating some evaluation metrices like: precision, recall, true relevant etc...

```python
def Evaluate_for_user(recommended,userId):
  fav = actual_favourite_for_user(userId)
  rec = get_top_recommended_N(10,recommended) 

  top_5_recommended = rec.head(5)
  top_10_recommended = rec.head(10)

  true_relevent = rec[rec['ISBN'].isin(fav)].shape[0]

  hit_at_5 = top_5_recommended[top_5_recommended['ISBN'].isin(fav)].shape[0]
  hit_at_10 = top_10_recommended[top_10_recommended['ISBN'].isin(fav)].shape[0]

  precision_at_5 = hit_at_5 / top_5_recommended.shape[0] if top_5_recommended.shape[0] != 0 else 1
  recall_at_5 = hit_at_5 / true_relevent if true_relevent != 0 else 1

  precision_at_10 = hit_at_10 / top_10_recommended.shape[0] if top_10_recommended.shape[0] != 0 else 1
  recall_at_10 = hit_at_10 / true_relevent if true_relevent != 0 else 1

  result = {'hits@5_count':hit_at_5, 
            'hits@10_count':hit_at_10,
            'recommended@5_count':top_5_recommended.shape[0], 
            'recommended@10_count':top_10_recommended.shape[0],  
            'relevents': true_relevent,
            'recall@5': recall_at_5,
            'recall@10': recall_at_10,
            'precision@5': precision_at_5,
            'precision@10': precision_at_10}

  return result
```

* Evaluate a model on all users

```python
def Debuuger(processed,total,prints):
  if processed/total >=0.25 and prints == 0:
    print('25% processed')
    prints = 1
  elif processed/total >=0.5 and prints == 1:
    print('50% processed')
    prints = 2
  elif processed/total >=0.75 and prints == 2:
    print('75% processed')
    prints = 3
  return prints
```

```python
def Evaluate(ModelName,ModelNumber):
  details = []
  i = 0
  L = list(set(test['User-ID'].values))
  print('Evaluating model on '+str(len(L))+' users')
  prints = 0
  for id in L:
    # DEBUG_START #
    # i+=1
    # print(i)
    # if i==500:
    #   break
    # DEBEG_END #
    i=i+1
    prints = Debuuger(i,len(L),prints)
    recommendations = get_Recommendations(ModelNumber,id)
    if len(recommendations) ==0:
      continue
    result = Evaluate_for_user(recommendations,id)
    result['user_id'] = id
    details.append(result)
    
    
  detailed_results_df = pd.DataFrame(details) \
                          .sort_values('hits@5_count', ascending=False)
      
  global_recall_at_5 = detailed_results_df['hits@5_count'].sum() / float(detailed_results_df['relevents'].sum())
  global_recall_at_10 = detailed_results_df['hits@10_count'].sum() / float(detailed_results_df['relevents'].sum())
  
  global_precision_at_5 = detailed_results_df['hits@5_count'].sum() / float(detailed_results_df['recommended@5_count'].sum())
  global_precision_at_10 = detailed_results_df['hits@10_count'].sum() / float(detailed_results_df['recommended@10_count'].sum())

  global_metrics = {'modelName': ModelName,
                    'recall@5': global_recall_at_5,
                    'recall@10': global_recall_at_10,
                    'precision@5': global_precision_at_5,
                    'precision@10': global_precision_at_10}    
  

  return global_metrics, detailed_results_df  
```
